### **9.5 - Password authentication and filter by owner**

As it should be the case for all of your applications running in production, restricting the access of

the Airflow UI is extremely recommended. In this video we are going to discover the different steps to follow

in order to set up the authentication system.

Let’s get started. [VS]From you code editor, check that you are under the section airflow-materials/airflow

-section-9 and open the file airflow.cfg in mnt/airflow. From there, look for the

parameter authenticate under the section web server.

As you may guess, this parameter is used to enable the authentication on Airflow. Meaning, accessing the UI

would be allowed only if you have a user account with a login password set. So defined the parameter

to true.

Then, we have to add a new parameter just below, in order to indicate the backend that will be requested

for authentication. Type: auth_backend

= airflow.contrib

.auth.backends

.password_auth

Like that. As you can see here, multiple backends are actually available with Airflow. If you copy this,

about one of these backends, don’t hesitate to reach me. All right,

back to the code editor. We can save the file and open the Dockerfile under the folder docker/airflow.

When we want to set up password authentication with Airflow, the passwords will be hashed by using the

module bcrypt. bcrypt is a password hashing function in order to encrypt your password in a very secure

manner.

I won’t go into the details here, since it's a bit technical, but if you are interested, don’t hesitate

to take a look at the bcrypt hash function. In order to make available this function, we have to install

a new package on the container running Airflow which is flask-bcrypt. So, just below a pip install

command,
