### **9.4 - [Practica] Hiding variables**

In this short video we are going to quickly discover a nice feature of Airflow, allowing you to hide

the value of some variables

according to the name you define.

Let’s get started. From you code editor,

check that you are under the folder airflow-materials/airflow-section-9 and open

the file airflow.cfg in mnt/airflow. From there,

look for the parameter

“hide_sensitive_variable_fields”.

This parameter is used to hide sensitive variable fields

when set to True which is the behavior by default.

So how can hide the value of a variable?

Well, open your web browser and go the Airflow UI by typing localhost:8080.

Enter. Click on “Admin”, “Variables” and “Create”. If we create random variable like “my_

var”

with the value “my_value”

then click on save, we obtain a new variable named my_var with the value my_value.

Notice that the value is actually encrypted as shown by the column here

using the fernet key we defined earlier meaning my_value will not be shown in clear from

the pathways database.

Nonetheless, if my_value would be a password, then anybody could see it just by accessing this

page.

One way to avoid this is my specifying one of many keywords given by Airflow to hide the values

of your variables.

For example,

if we create a new variable with the name “my_password”

and a random value, like that. Click on save.

This time the value is hidden as shown here.

Why?

Because in the name of the variable we specified the keyword password.

All right, last thing before moving forward.

This security is actually purely cosmetic. Meaning, if a given DAG gets the variable and displays

the value, then potentially anybody would be able to read that value.

One way to fix this would be to encrypt the value of your variable with a key and then, once the variable

has reached its final destination, use this key to decrypt the value.
