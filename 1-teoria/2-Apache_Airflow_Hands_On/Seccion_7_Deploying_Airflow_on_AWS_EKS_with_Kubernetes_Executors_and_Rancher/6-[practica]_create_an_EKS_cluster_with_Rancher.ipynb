### **7.6 - [Practica] Create an EKS cluster with Rancher**

Now all the stuff with AWS is set up

we can create an EKS cluster with Rancher.

If you have seen the previous videos in section 5, you should already know this interface since we used

it in order to import a local kubernetes cluster.

This time, instead of importing a cluster, we gonna create one and let Rancher does the hard

stuff for us.

So if you are ready, click on “Add cluster”. From there,

Rancher gives you many possibilities according to the kind of kubernetes cluster you have or you

want.

For example, you can create a cluster from existing nodes or by using the Rancher Kubernetes Engine

which is a certified Kubernetes distribution that runs entirely within Docker containers.

It works on bare-metal and virtualized servers.

It’s purpose is to solve the problem of installation complexity which you may already know if you ever

tried to install Kubernetes by yourself.

It’s a nice distribution so I strongly recommend you to take a look at it.

Another option for creating clusters is by choosing between the available hosted kubernetes providers.

Select Amazon EKS. From there,

you have to give some information to set up you cluster.

Don’t worry it’s gonna be very easy.

First, for the cluster name we type “airflowEKScluster”.

Then, nothing to change for the member roles,

labels and annotations. In the account access panel,

this is where you need to specify the security credentials from the csv file you downloaded earlier.

For the region,

select the same than the one you set for your aws console. To check the region,

if you go on your AWS console,

at the top right corner, you have the nearest city from your location.

In my case it’s Paris but obviously that’s can be different for you.

Click on the city and next to it, you have the region.

In my case it’s eu-west

-3.

Again, you should specify your region and not mine. Back to Rancher, select the correct region.

Now we need to input the credentials. Open your credentials

.csv file. Copy the access key ID

and paste it here. Then copy the secret access key

and put it there.

Perfect. Click on “Configure Cluster”. Here

you can choose the Kubernetes version you want and ultimately the most recent supported kubernetes

version.

In my case it’s 1.14. We keep the default value for the service role. Next.

Same for the VPC and subnet, we let Rancher does the work.

Again nothing to change.

We choose the instance type we want, which are nothing more than EC2 instance types. We can specify

the number of nodes, the node volume size and so on.

. For now, we keep everything as set by default and click on “Create”. Ok

at this time, it will take 10 to 15 minutes for the EKS cluster to be created.

In the meantime, you can see the cluster status from the AWS console.

Indeed, Rancher kicks off the installation by using CloudFormation templates

so if you look for the service “cloudformation” from the services,

you can watch the status of the different stacks being created in real time. Also,

if you take a look into the EKS service,

you can follow the status of your EKS cluster. At the end of the process you should get the status

active from the Rancher UI next to the name of your cluster.

So I’m gonna pause the video right now and come back when it’s done.

Welcome back,

so after a long time the cluster is finally active.

If we click on it, we land on a dashboard giving us some basic information about the cluster

such as CPU and memory usage,

how many pods are running, the status of the different Kubernetes components and so on.

Just below, you have the events of the cluster which can be a great help to troubleshoot some errors.

One very important button is the one here to launch a kubectl session.

If you click on it, you get connected to your cluster and can execute any kubectl commands.

For example, we can type “kubectl version” to display the kubernetes version running on the client and server.

Notice that the versions are not the same since only the version 1.14 is supported on EKS at this

time.

Okay if we type “kubectl get nodes”,

we get the list of nodes.

For now we only have one node.

The command “kubectl get pod

--all-

namespaces”

allows to list all the running pods in the cluster, and there are many other commands that you can

use with kubectl but we will focus on the ones we need.

Alright we can close the terminal and now you have a beautiful EKS cluster set up and running on

AWS.

Let's take a quick break and see you in the next video.
