### **7.2 - Quick overview of AWS EKS**

In this video I’m going to give you a quick overview of AWS EKS. You should already be

familiar with Docker and Kubernetes so I won’t talk about them again.

If not, I strongly recommend you to take a look at the videos “Docker Reminder” in section 3 and “Kubernetes

Reminder” in section 5.

That being said let's move forward.

So what is AWS EKS? If you already use AWS services,

you may have heard about AWS ECS for Elastic Container Service.

AWS ECS came before AWS EKS and is a provider-specific solution for running

and managing docker containers in the cloud. While Kubernetes is an open-sourced container orchestrator,

Amazon made the choice to go with a proprietary solution which is AWS ECS but something was

missing. In 2017 AWS announced the support of Kubernetes with AWS

EKS for Elastic Kubernetes Service. EKS is a managed service for Kubernetes, where users can easily

deploy, a high-available, scalable and fully managed Kubernetes control-plane on AWS. Like with

ECS, the goal of EKS is to make your life easier by managing what cause you a lot of pain

when you start dealing with Kubernetes. EKS is well integrated now in AWS and can be used with

different services such as Amazon ECR for container images, Elastic Load Balancing or ELB for load

distribution,

IAM for authentication and so on.

Let me show you the kind of Kubernetes architecture we usually want.

Here, we have three availability zones A, B and C in order to maintain a high availability.

Then for each zone, we would need to set up a master node, an etcd instance which is the datastore

of Kubernetes and once everything is setup, we could start deploying kubernetes worker nodes.

So, basically you need to manage by your own, three master nodes,

three etcd instances and many kubernetes worker nodes all spread among different availability zones.

The good news is, a lot of things can fail and it will be up to you to do the work.

So before EKS, if you really wanted to set up your own Kubernetes cluster in AWS you would

have made a lot of effort.

Now things are easier since EKS manages your master nodes and etcd instances for you. It will automatically

scale and repair your services if needed.

More specifically, it runs the control plane instances

across the zones and detects and replaces unhealthy control plane instances. If you don’t remember

what is a control plane in Kubernetes, it’s where you have your etcd instance, kube-controller, kube

-apiserver and so on running.

Basically it’s everything you don’t want to deal with.

Now, even if EKS is in charge of the biggest part, you still need to create the kubernetes worker nodes

along with IAM roles, VPCs, Subnets, SSH keys and so on.

We will see that even this part can be easier to set up by using a tool like Rancher.
