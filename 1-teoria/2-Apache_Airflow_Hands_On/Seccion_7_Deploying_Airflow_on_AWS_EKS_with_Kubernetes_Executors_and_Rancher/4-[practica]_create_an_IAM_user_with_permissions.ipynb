### **7.4 - [Practica] Create an IAM User with permissions**

In this video we are going to create a new user as it is required by Rancher to set up an EKS cluster.

From you AWS console,

click on “Services” and type “IAM”.

Select it. then go to “Users” and “Add users”.

From there,

we have to fill in some information about the user we want to add. The first thing to give is the username.

In my case, I’m gonna type “airflow-user”.

Then we have to select how the user will access AWS.

Since Rancher asks for an access key ID and a secret access key, select the first option “Programmatic

access” and move forward. Here, we can set the different permissions to our new user.

I already have some groups which may not be your case but don’t worry you don’t need them

since we are going to attach existing policies directly to the user.

Nonetheless, keep in mind that you should create roles and groups with the correct policies depending

on the access you want to give in production.

For now we are going to stay very simple.

Click on “Attach existing policies directly”. As you can see from there, AWS gives you many policies

to apply. In our case,

we gonna give the administrator policy to the new user so that it will be able to access to

everything.

Again, it’s really just for testing purposes and it’s absolutely not what you should do in production.

If you want to have more content about it, don’t hesitate to tell me in the Q/A section.

So, search for “Admin” here

and select “AdministratorAccess”.

Then click on Next.

We don’t have any tags to specify. “Review”, ok,

everything seems to be good.

We can create the user.

Perfect,

now the user is created,

you absolutely need to download the .csv file that contains the security credentials.

Don’t lose this file as you won’t be able to download it again.

Finally, we close the window and we can find airflow-user from the Users

panel.

Good,

that’s it for this part,

let’s take a quick break and see you in the next video.
