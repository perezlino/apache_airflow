### **7.5 - [Practica] Create an ECR Repository**

In the previous video we have set up an EC2 instance running Rancher and created a new user with

the permissions on AWS services.

In this video we are going to configure an ECR repository in order to store, manage and deploy Docker

container images. An ECR can be useful in the context of a CI/CD pipeline where you will push

your image containing your DAGs and update your Airflow instance with it.

Alright, let’s begin. From you AWS console,

look for the service ECR

and click on “Get started”. One important thing to mention is that ECR is not free.

The cost is quite cheap like $0.09 cts per gigabyte transferred but it’s good to know that you

can be charged for using this service.

Okay here,

the first thing to do is to give a name.

Let’s put “airflow”.

Notice that you can set a namespace before the name. It’s a good practice so that you could divide

your repositories by environments such as dev, staging and prod.

That being said, we can click on “create repository”, and that’s it.

The repository airflow is created, we are able to push docker images into it.

So, in order to push the docker images we need to install the AWS command line interface. Open a

new tab and search for “installing AWS

CLI 2”.

Click on the first result

and at the bottom of the page, you can choose between links according to your operating system.

terminal,

paste the commands to execute them.
