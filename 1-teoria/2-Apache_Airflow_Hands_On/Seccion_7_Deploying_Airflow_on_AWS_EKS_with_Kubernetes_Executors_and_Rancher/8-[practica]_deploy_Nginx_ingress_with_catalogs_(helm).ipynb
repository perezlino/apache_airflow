### **7.8 - [Practica] Deploy Nginx Ingress with Catalogs (Helm)**

In this video we are going to install the Nginx Ingress in our Kubernetes cluster.

Fortunately for us, installing Nginx Ingress on EKS

with Rancher, is an easy task

thanks to the Catalogs. A catalog in Rancher is nothing more than a set of Helm charts in order to deploy

complex applications without having to configure everything.

It’s a nice way to avoid creating many yaml files and suffering with mysterious errors because you

try to install everything by yourself.

Well, actually some people did and that’s why it’s easier now to deploy some applications.

Ok enough talking let’s move on. From the Rancher user interface,

click on “airflowekscluster”, select “global” and click on “Apps”.

Here, we don’t have any apps running yet

but first let’s turn on the catalog corresponding to the official helm repository.

Click on “Manage catalogs” and by clicking on this button next to the helm catalog, we can enable it.

As you can see, it turned active and begins to synchronize all the helm charts with Rancher.

So let’s wait a few minutes until the synchronization is done.

Okay now it's finished.

If we go to “Apps” again and click on “Launch”, we can see all applications that we can install automatically

using catalogs.

There is grafana, efk if you want the elastic stack, hadoop, kafka and so on.

Notice that each subset of catalogs is divided by the repository where it comes from.

So, “library” is the native catalogs of Rancher

but if continue to scroll down, you will get the helm charts. Ok, let’s find Nginx ingress. Type “nginx”

and select “nginx ingress”. From there, you can get a detailed description of the application you want

to install as well as some options to configure.

First, we have to define a name, let’s say “nginx-ingress”.

Then you can select the template version you want. By default Rancher will select the most recent one. “Target

projects” is where you want to run the app. We choose the default project under “airflowekscluster”. Here,

there are two options for your upgrade strategies either by rolling update or by updating all apps simultaneously.

We stick with the default value. Finally, in “available roles”, we select cluster so that the app can access

any resources in the cluster. “Answers”

here, is where you can overwrite the default values in the values.yml file. Most charts

include a file called values.yaml which provides default configuration data for the templates in

a structured format.

You can take a look at this file by clicking on preview right there.

And as you can see, there are different keys and values that you can overwrite through answers.

Alright, let’s click on “Launch”. The app is being installed,

let's click on it.

This page shows you an overview of the components your app deployed in your kubernetes cluster.

You can check the endpoints, workloads, ingress rules, services, volumes and so on.

For example, if we click on the nginx-controller, we access to the deployment object and get more

details about the running ports,

events, environment variables set and so on.

It’s also possible to check the logs of a specific pod by clicking here, then “view logs”,

or, get an access to the container by clicking on “Execute shell”.

Back to the app dashboard,

go to the services and just under the nginx controller service,

you can find the port 443 and 80. By clicking on port 80 for example, we get an access to the nginx

backend running inside our kubernetes cluster from the outside.

This message right here is being created by the default backend created for us by the nginx ingress.

It also provides a default page when someone tries to access a route which doesn’t exist inside of our

application.

Alright, now the ingress is ready, and we can access to our cluster from the outside,

let’s move to the next video where we gonna install Airflow inside our EKS cluster.
