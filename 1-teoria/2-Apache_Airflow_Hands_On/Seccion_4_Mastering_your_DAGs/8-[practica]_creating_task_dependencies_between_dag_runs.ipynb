#### **`wait_for_downstream`**

Agregamos a la tarea **`bash_task_1`** la dependencia **`wait_for_downstream=True`**

<center><img src="https://i.postimg.cc/mkkQdtQh/a675.png"></center>

Entendiendo que el Schedule_interval está configurado para ejecutarse cada medianoche a partir del 01/01/2019:

<center><img src="https://i.postimg.cc/0Q8GwtjY/a676.png"></center>

El primer DAG Run efectivamente ejecutado ocurre el 02/01/2019 a las 00:00 : 

<center><img src="https://i.postimg.cc/3Nq2zYB1/a677.png"></center>

En la Tree View vemos que las tareas se ejecutaron correctamente:

<center><img src="https://i.postimg.cc/7Pt15nkz/a678.png"></center>

Ahora, vamos a plantear una excepción de la función second_task:

<center><img src="https://i.postimg.cc/dtz8BQDt/a679.png"></center>

Como era de esperar, tenemos la segunda tarea con error y la tercera con el estado upstream_failed. Nada nuevo. 

<center><img src="https://i.postimg.cc/vH1rfxYN/a680.png"></center>

Ahora volvemos a dejar las funciones como estaban en un principio y volvemos a ejecutar nuestro DAG:

<center><img src="https://i.postimg.cc/k41StqkC/a681.png"></center>

En el nuevo DAG Run ninguna de las tareas se activará. ¿Qué ha pasado aquí? Bueno, como la segunda tarea python_task_2 ha fallado en el DagRun anterior y porque está inmediatamente después de la tarea bash_task_1, esa misma tarea no se ha ejecutado en el DagRun actual. En realidad, ninguna de las tareas se ha ejecutado porque la primera tarea que se dispara en el DAG es bash _task_1. Si queremos que el DagRun se ejecute, sólo tenemos que cambiar el estado de la segunda tarea en el DagRun anterior de failed a success.

<center><img src="https://i.postimg.cc/52L8kcF5/a682.png"></center>

**Como volver a ejecutar la tarea fallida usando “Clear”**

En realidad, sólo para mostrar cómo se puede volver a ejecutar una tarea fallida, vamos a borrar (clear) su estado de la metadatabase. Para ello, haz clic en la tarea aquí y haz clic en "Clear". Como puede ver en el mensaje de arriba, 2 tareas han sido borradas, lo que significa que están listas para ser ejecutadas. Actualice la página varias veces y verás que tus DagRuns se están completando. Y ya está hecho. Porque la segunda tarea del DagRun anterior ha tenido éxito, entonces el DagRun actual ha podido activar la primera tarea con el parámetro wait_for_downstream.

1.	Damos clic en la tarea fallida:

<center><img src="https://i.postimg.cc/43GVD8zD/a683.png"></center>

2.	Pulsamos CLEAR

<center><img src="https://i.postimg.cc/k4X84BYw/a684.png"></center>
<center><img src="https://i.postimg.cc/HLtXdQzV/a685.png"></center>

Dado que habíamos modificado el código para que ninguna tarea devolviera un error, de esta manera con CLEAR nos permite borrar los resultados fallidos obtenidos y así volver a re ejecutar desde la tarea fallida en adelante.
