### **6.6 - Trigger rules for your tasks**

- **`all_success`**
- **`all_failed`**
- **`all_done`**
- **`one_failed`**
- **`one_success`**
- **`none_failed`**
- **`none_skipped`**

Cuando creas tus DAGs tienes que definir dependencias entre las tareas que quieres ejecutar.  Por defecto, cada tarea se ejecuta después de que sus padres directos hayan tenido éxito.  Ese es el comportamiento por defecto. En este vídeo, vamos a descubrir cómo puede crear dependencias más complejas entre sus tareas con trigger rules.  Empecemos con un ejemplo.  

<center><img src="https://i.postimg.cc/6p068n81/a965.png"></center>

Este DAG describe un data pipeline en el que se descargan datos de dos sitios web diferentes.  Luego los datos son procesados y combinados (merge) por la tarea en rojo.  En este punto, es posible que desee elegir la siguiente tarea a ejecutar en función del resultado de la tarea merge.  En nuestro caso, si la tarea falla, nos gustaría ejecutar la tarea slack_notification, mientras que, si la tarea tiene éxito, nos gustaría ejecutar la tarea email_notification.  Eso es lo que las trigger rules te permiten hacer y puedes definirlas estableciendo el parámetro "trigger _rule" en tus operadores.  Hay muchas trigger rules diferentes, así que déjame mostrarte cada una junto con un ejemplo.  

En primer lugar, tenemos all_success que es la trigger rule por defecto definida para sus tareas.  Básicamente, aquí, si las tres tareas upstream de t4 han tenido éxito, entonces t4 se dispara.  

<center><img src="https://i.postimg.cc/VNkfnvMs/a966.png"></center>

El ejemplo opuesto, es si una de las tres tareas upstream ha fallado, entonces t4 se establece en "upstream_failed" estado como se muestra aquí. Por cierto, fíjate que he mantenido los mismos colores que Airflow, naranja para "upstream_failed", verde oscuro para "success", rojo para "failed" y así sucesivamente.  La siguiente trigger rule es "all_failed". 

<center><img src="https://i.postimg.cc/g0tYDpqV/a967.png"></center>

A diferencia de "all_success" t5 se activará sólo si sus tareas upstream, t2, t3 y t4, están en estado fallido o upstream_failed.  

<center><img src="https://i.postimg.cc/9X7WGv49/a968.png"></center>

Si uno de los padres ha tenido éxito, entonces la tarea t5 con la trigger rule "all_failed" será omitida como se muestra aquí.  

<center><img src="https://i.postimg.cc/vHwGX39N/a969.png"></center>

La tercera trigger rule que depende de todos los padres de una tarea es "all_done".  Como puedes adivinar por su nombre, esta regla disparará tu tarea cuando cada tarea upstream haya sido ejecutada sea cual sea su estado. Aquí, t1 ha fallado, t2 tiene el estado upstream_failed, y t3 ha tenido éxito.  No obstante, con la regla all_done, se dispara la tarea t4. t4 no se disparará mientras una de sus tareas upstream siga en ejecución como se muestra en este ejemplo.  

<center><img src="https://i.postimg.cc/hjHck0Yh/a970.png"></center>

Muy bien, en algún momento es posible que desee activar una tarea tan pronto como al menos una tarea upstream haya tenido éxito o haya fallado.  Eso es lo que puedes hacer con las trigger rules one_success y one_failed. 

<center><img src="https://i.postimg.cc/DyQnNLPy/a971.png"></center>

En este caso, t3 sigue esperando a ser ejecutada, pero como t1 ya ha fallado, se activa t4.  Tenga en cuenta que la tarea con la trigger rule one_failed o one_success se ejecutará sin esperar a que sus padres terminen, siempre y cuando al menos uno de los padres haya fallado o tenido éxito. t4 no se disparará y se omitirá si todas las tareas upstream han tenido éxito. 

<center><img src="https://i.postimg.cc/j5WR31ZG/a972.png"></center>

Para "one_success" es exactamente el mismo comportamiento, pero esta vez t4 se dispara tan pronto como uno de los padres ha tenido éxito. 

<center><img src="https://i.postimg.cc/Znp44dkH/a973.png"></center>

Y t4 no se disparará si todos los padres han fallado como se muestra en este DAG.  

<center><img src="https://i.postimg.cc/FH9NXzTb/a974.png"></center>

Muy bien, hay dos últimas trigger rules que quiero mostrarte.  La primera es "none_failed".  

<center><img src="https://i.postimg.cc/3xn78hqs/a975.png"></center>

En este caso, t4 se ejecutará sólo si todos los padres no han fallado.  Por ejemplo, si un padre ha tenido éxito y otro se ha omitido, entonces t4 será capaz de ejecutar.  Pero, si alguno de los padres está en estado fallido o upstream_failed, entonces t4 también estará en estado upstream_failed.

<center><img src="https://i.postimg.cc/SKR4j5FJ/a976.png"></center>

La última trigger rule es "none_skipped", lo que significa que, si ningún padre se encuentra en estado omitido, se activará t4. 

<center><img src="https://i.postimg.cc/ZqQSNc4j/a977.png"></center>

De lo contrario, si uno de los padres se ha saltado, entonces t4 se establecerá con el estado upstream_failed.  

<center><img src="https://i.postimg.cc/j5Jr47jT/a978.png"></center>

Bien, no dudes en echar un vistazo a los diferentes ejemplos que hemos visto para acostumbrarte a las trigger rules ya que son muy útiles. Como nota final, debes saber que hay una última trigger rule llamada "dummy".  Puede utilizar el parámetro depends_on_past con las trigger rules para controlar aún más cómo se ejecutan sus tareas.  Por último, las tareas omitidas se ejecutarán en cascada a través de las trigger rules all_success y all_ failed.  Esto significa que si tienes una tarea con la trigger rule all_success o all _failed y uno de sus padres ha sido omitido, entonces esa tarea también será omitida a menos que cambies su trigger rule por all_done, one_success, one_failed y así sucesivamente.  Cualquier otra regla excepto all_success y all_failed.  Muy bien, ahora es el momento de ver las trigger rules en acción. Nos vemos en el próximo vídeo.
