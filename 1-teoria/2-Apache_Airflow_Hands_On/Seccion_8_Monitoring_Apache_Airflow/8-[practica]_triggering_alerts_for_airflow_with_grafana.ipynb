### **8.8 - [Practica] Triggering alerts for Airflow with Grafana**

In this video I would like to show how you can you set up alerts with Grafana in order to monitor

your Airflow instance and be warned automatically when something goes wrong. In order to make all Airflow

measurements available in Grafana,

we actually have to trigger a DAG.

Indeed, if you go to the metrics of Airflow from the documentation and take a look under the section

Timers. For example, the metrics depending on a dag id or task id, here, usually need to have the corresponding

DAG triggered at least one time in order to make the metric available.

So, in the Airflow UI, click on this little button to trigger the DAG logger_dag and go back to Grafana.

Click on “add panel”, then “add query”, “select measurements”, and as you can see we have much more metrics than

in the previous video.

Why?

Because by triggering a DAG, Airflow was able to send all the metrics related to DAGs execution such

as the milliseconds taken to finish a task, the overall task instances successes and so on.

All right, since we are going to create an alert from Grafana that will be sent through Gmail, you

have to set up your account like we did in the video

“Sending an email notification” from the section 3. If you didn’t watch it, pause the video right now and

follow the different steps in order to obtain the password needed for sending an email from your Gmail

account.

check that you are under the folder airflow-materials/airflow-section-8 and open

the file grafana.ini in mnt/grafana. Look for the “smtp” section,

and uncomment the following lines.

All right,

now the channel is defined,

let’s create a new panel in the Airflow dashboard.

Go to “Dashboard”, “Airflow”, “Add panel” and “Add query”. Select the measurement “airflow_dag_logger

_dag_t2_duration”.

Click on “value” here and select 100_percentile corresponding the duration.

Click on “mean”

then “remove”.

Click on the “plus”, “aggregations” and select “distinct”. As you can see, a point appeared just there giving

the time took by the task t2 to finish in milliseconds.

Actually, let’s run the DAG two more times so that we will have more data.

Ok, refresh the dashboard, and we obtain two new durations.

Now, click on “Visualization” from the left. Then, under the section “Axes”,

click on the value of “Unit”. Scroll down until you reach “Time”.

Click on it, and select duration in seconds in order to express the duration at most in seconds from

the Y axis. Move to “General” and set the title to “logger_

dag.t2 duration”.

Good. Last thing,

click on the bell here, then “Create Alert”.

From there, we can create an alert that will be checked every 1 minute and you won’t notified twice

in an interval of five minutes.

Click on average here, and select last.

Then after “is above”, type 5 so that an alert will be triggered if the last task took more than 5

seconds to finish in the last five minutes as defined here.

Notice that “No data and error handling” are enabled

and so you will receive an alert as well

if you don’t receive any data in the last 5 minutes or if you got an error. Next, under the section

“notifications“ click here, and select “email-channel”. Then in message type “t2 in logger_

dag took more than 5 seconds to finish”.

Now the alert is set up.

Click on “save dashboard” from the top bar. “Save” and go back to the dashboard. From the top right corner,

click here, and select 5 seconds so that the dashboard will be refreshed every 5 seconds automatically.

Finally it’s time to verify that the alert we configured works. To do this,

go to your code editor and in the folder mnt/airflow/dags,

open the file logger_dag.py.

As you can see, t2 only prints out a message on the standard output which is very fast to execute.

Let’s remove this command, and type “sleep 10” instead.

By doing that, the task t2 should take more than 5 seconds to finish and so the alert should be triggered.

Save the file and go the Airflow UI. Click on logger_dag, then refresh the DAG by clicking

here,

and trigger it. “Ok”.

Now open your dashboard and wait for the event to appear on the chart.

Here it is.

As you can see the task took more than 5 seconds to finish and crossed above the red line, meaning

an alert has been sent.

If we check yopmail again, and click on “check for new mails”, here we got the alert.

Perfect.

So now you are able to monitor your Airflow instance as well as receiving alerts from Grafana to be

warned if anything goes wrong. Well done! By the way,

if you received an email starting with No Data instead of the alert we set, it means that Grafana

didn’t receive any data from the DAG logger_dag

In the past five minutes.

One important thing to keep in mind is that you will not receive two notifications in an interval of

five minutes,

so if the no data alert has been sent, then you won’t receive the alert about the task t2 taking more

than 5 seconds to finish. The no data alert is sent in priory and doesn’t wait for the interval

of five minutes.

So to sum up, keep executing your DAG every minute, and it will work.
