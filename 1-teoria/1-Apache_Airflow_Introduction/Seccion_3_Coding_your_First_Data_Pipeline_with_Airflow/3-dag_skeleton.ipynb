#### **`start_date, schedule_interval y execution_date`**

A continuación, hay que definir la fecha de inicio (**`start_date`**). La fecha de inicio define la fecha en la que su DAG comienza a ser programado (scheduled). Para definir esta fecha necesitas importar "datetime" y definimos la fecha el 1 de enero de 2022.

<center><img src="https://i.postimg.cc/FzVBRk5x/a97.png"></center>

Una vez que tiene la fecha de inicio, necesita definir el intervalo de programación (**`schedule interval`**), **la frecuencia con la que su DAG se dispara** una vez al día, cada 15 minutos, una vez a la semana y así sucesivamente, con el valor idealmente diario que podemos establecer. Este DAG se dispara todos los días a medianoche.
```
Si no indicamos el schedule interval se establece en un día. Así que por defecto, sus DAGs están programados (scheduled) para ejecutarse cada 24 horas. Ese es el schedule interval por defecto.
```
Hay que tener en cuenta que el intervalo de serie se define como una expresión Cron. Así que detrás del @daily tienes una expresión Cron. Puedes encontrar estas anotaciones en la documentación de Airflow.

<center><img src="https://i.postimg.cc/XNfhxmGy/a98.png"></center>

**`Su DAG se dispara (triggered) efectivamente una vez que la fecha de inicio (start date) más el schedule interval han transcurrido`**. Por ejemplo, si tiene una start date, digamos el 1 de enero de 2021 a las 10:00 a.m. y el schedule interval se establece en 10 minutos. Su DAG se activará efectivamente el 1 de enero de 2021 a las 10:10 a.m. Su DAG se dispara (is triggered) efectivamente después de la start date más el schedule interval.

**==> ¿Qué ocurre en Airflow?**

La start date es el 1 de enero de 2021 a las 10:00 horas y el schedule interval es de 10 minutos. 

**==> Entonces, ¿qué ocurre?** 

Todavía nada. De hecho, Airflow espera 10 minutos y no es hasta las 10:10 horas cuando se crea y se ejecuta el DAG Run. En cuanto el DAG Run se completa, el 1 de enero de 2021 a las 10:00 horas se convierte en la "**`Execution date`**". La "Execution date" es el inicio de este periodo, del scheduled period. Y entonces las 10:10 a.m. se convierte en la "Start date".

**==> ¿Qué es lo que sigue?**

Pues bien, de nuevo, Airflow espera 10 minutos y, a continuación, a las 10:20 horas, se crea y se ejecuta el DAG Run, el segundo DAG Run, el DAG Run 2. En cuanto se completa el DAG Run 2, la "Execution date" para ese DAG Run específico, el DAG Run 2, pasa a ser el 1 de enero de 2021 a las 10:10 a.m. Y la nueva Start date es las 10:20 a.m.
```
Así es como funciona. Recuerda que cada DAG Run tiene una execution date y esta execution date corresponde al inicio del scheduled period para ese DAG Run específico.
```

#### **`catchup`**

El último parámetro que queremos añadir aquí es el parámetro **`catchup`**. **Te aconsejo encarecidamente que lo pongas en `false`**.

**¿Por qué?**

**Porque por defecto el parámetro `catchup` está establecido en `true`**. Y significa que si entre ahora y la fecha de inicio (start_date) su DAG no ha sido disparado, entonces tan pronto como usted comience a programar su data pipeline desde la Airflow UI, usted va a catchup (catchup), es decir, se ejecutarán todos los DAG runs "non triggered" entre este período de tiempo y la fecha de inicio o la última vez que se activó el DAG.

Así que en mi opinión, siempre es bueno establecer el parámetro catchup a false para que pueda controlar exactamente que DagRuns quieres ejecutar y no va a terminar con una tonelada de DAG runs que se ejecutan al mismo tiempo.

No te preocupes, lo verás más adelante en el curso.

#### **`Diferencia entre timedelta y cronn`**

**`¿Cuál es la diferencia entre estos objetos timedelta y la expresión cron?`**

Una expresión cronn es absoluta, lo que significa que si dices que quieres disparar tu DAG todos los días a las 7:00 P.M., siempre ejecutarás tu DAG todos los días a las 7:00 P.M. Mientras que con el objeto timedelta, este schedule interval se transmite a ti. Por ejemplo, si con el objeto timedelta estableces "hours=7" tu DAG se ejecutará cada siete horas a partir de la execution date anterior.

Así que con una expresión cron, definiste que quieres disparar tu DAG en cada intervalo de tiempo específico. Mientras que con el objeto timedelta defines que quieres disparar tu DAG cada intervalo de tiempo determinado. Así que el objeto timedelta es un schedule interval relativo.

<center><img src="https://i.postimg.cc/qvhxD6bZ/a121.png"></center>

Por ejemplo, utilizando una expresión cron nuestro objetivo es disparar nuestro DAG cada tres días. Si nos fijamos que hay un error respecto al último día de Febrero, no existe una diferencia de tres días, sino, de 1 día:

Al definir el valor del schedule interval a "**`None`**", esto significa que tu DAG nunca será disparado automáticamente por el scheduler. Y esto es realmente útil cuando usted quiere disparar su DAG sólo manualmente, o por su trigger externo.
