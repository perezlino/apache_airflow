### **5.3 - How to use SubDAGs**

Imaginemos que tienes un dag con estas tareas:

<center><img src="https://i.postimg.cc/CxjgLW0V/a412.png"></center>

download_a, download_b y download_c y transform_a, transform_b, y transform_c. Lo que te gustaría hacer es agrupar esas tareas, como una tarea downloads para download_a, download_b, y download_c, y una tarea transforms para transform_a, transform_b, y transform_c.

**¿Cómo se puede hacer eso?**

Permítame mostrarle esto con **`SubDAGs`**.

El primer paso es ir a la carpeta "dags", crear una nueva carpeta en ella, llamarla "subdags", y luego crear un nuevo archivo llamado subdag_downloads.py.

<center><img src="https://i.postimg.cc/1X2QfS4y/a413.png"></center>

Recuerda que el SubDAG no es más que un DAG con las tareas que quieres agrupar.

Así que el primer paso es importar el DAG object. Luego hay que importar el BashOperator como la tarea que se quiere agrupar, usa el BashOperator y luego hay que definir una función. Esa función devuelve un DAG object.

<center><img src="https://i.postimg.cc/jdnTBDk2/a414.png"></center>

Pero primero vamos a definir los parámetros. El primero es el "parent_dag_id". El SubDAG debe tener el parent_dag_id en su DAG ID. Luego tienes el "child_dag_id" que corresponde al DAG ID del SubDAG y luego los argumentos. Así que hay que asegurarse de que algunos argumentos son los mismos entre el SubDAG y el Parent DAG, como como el schedule interval, the start date o el capture parameter. De lo contrario, puedes terminar con comportamientos extraños.

<center><img src="https://i.postimg.cc/2yPsCxX6/a415.png"></center>

A continuación, se define el DAG object como se hace habitualmente para definir un DAG y se especifica el DAG ID del SubDAG. Como puedes ver, hay una notación especial aquí. Debes poner primero el "parent_dag_id" con un punto y luego el "child_dag_id" que corresponde a tu SubDAG DAG ID. Esto es muy importante, de lo contrario no funciona.

<center><img src="https://i.postimg.cc/bw07jk0y/a416.png"></center>

A continuación defines el objeto DAG y pones la primera tarea que quieres agrupar, que es la descarga_a, algo para la descarga_b y algo para la descarga_c. Una vez que tienes todas las tareas que quieres agrupar, estás listo para devolver el DAG al final de esa función. Eso es todo.

<center><img src="https://i.postimg.cc/k5nkMGwx/a417.png"></center>
<center><img src="https://i.postimg.cc/yNv2DKr2/a418.png"></center>

En este punto has creado con éxito una función "subdag_downloads" que devuelve un DAG, tu SubDAG con las tareas que quieres agrupar. Y este es realmente el primer paso para crear un SubDAG. El siguiente paso es ir a tu DAG "group_dag" y luego tienes que hacer la primera importación, que es el "SubDagOperator". El "SubDagOperator" es el encargado de llamar al SubDAG que acabas de crear. Pero primero tienes que definir el "objeto args" con los parámetros de fecha de inicio, intervalo de programación y catchup del Parent DAG que vas a utilizar en tu SubDAG.

<center><img src="https://i.postimg.cc/3RNVpqsN/a419.png"></center>

Y finalmente puedes eliminar las tres tareas, download_a, download_b y download_c, y sustituirlas por el SubDagOperator. Como para cualquier otra tarea se define un task_id = "downloads" y luego se define el subdag. El subdag espera un DAG, por lo que utilizas la función subdag_downloads que has creado. Pasas el "dag_id" del Parent DAG con "dag_id" y pasas "downloads" como el Child ID. Es realmente importante que este child DAG ID sea el mismo que el definido en el task ID. Así que el task ID del SubDagOperator debe ser usado como el DAG ID de su subdag. Es realmente importante. De lo contrario, no funciona. Recibirás un error.

<center><img src="https://i.postimg.cc/W3yK7nVn/a420.png"></center>
<center><img src="https://i.postimg.cc/5yFrKCNM/a421.png"></center>

Luego pasas los argumentos "args". Finalmente importas la función "subdag_downloads"

<center><img src="https://i.postimg.cc/pyb6wLp5/a422.png"></center>

y cambias las dependencias para usar tu SubDagOperator, en lugar de esas tres tareas usas "downloads". 

<center><img src="https://i.postimg.cc/PfprnVCd/a423.png"></center>
<center><img src="https://i.postimg.cc/4xZJtJB3/a424.png"></center>

Guardas el archivo y luego puedes comprobar en la Airflow UI lo que obtienes, refrescas la página y como puedes ver, terminas con una tarea "downloads".

<center><img src="https://i.postimg.cc/PfgfXbDY/a425.png"></center>

Si disparas (trigger) tu DAG

<center><img src="https://i.postimg.cc/jdyxLW69/a426.png"></center>

haga clic en downloads > Zoom into SubDAG > Graph View.

<center><img src="https://i.postimg.cc/L8Tm3rb3/a427.png"></center>
<center><img src="https://i.postimg.cc/pdLRGsNb/a428.png"></center>

Puedes ver las tres tareas download_a, download_b y download_c que están bien agrupadas.

<center><img src="https://i.postimg.cc/bvV8K18G/a429.png"></center>

Así es como puedes agrupar tus tareas y hacer que tu DAG sea más fácil de leer con Sub Dags. Recuerda que para usar un SubDAG necesitas importar el SubDagOperator. Luego necesitas definir el task ID del SubDagOperator como parte del DAG ID del subdag. Entonces necesitas crear una función que devuelva un DAG correspondiente a tu sub dag. Tienes que asegurarte de que los argumentos entre el Parent Dag y el subdag son los mismos, como el parámetro catchup, el schedule interval y la start date. Por último, pero no menos importante, recuerda la notación especial para el SubDAG ID Parent DAG ID punto el Child DAG ID que corresponde al task ID del SubDagOperator.
