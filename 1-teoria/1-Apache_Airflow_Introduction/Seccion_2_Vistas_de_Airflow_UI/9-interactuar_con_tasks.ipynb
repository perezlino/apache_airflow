### **2.9 - Interactuar con Tasks**

- ¿Cómo puedo interactuar con las tareas?
- ¿Cómo puedo obtener los logs de una tarea determinada?










Imaginemos que estás en la Graph View, pero esto funciona tanto en Tree View como en Gantt View. Así que si haces clic en "batch_print_date1" por ejemplo:

<center><img src="https://i.postimg.cc/MGVJFy69/a70.png"></center>

Se llega al menú contextual de la instancia de la tarea (task instance context menu):

<center><img src="https://i.postimg.cc/ZYVt2Ymz/a71.png"></center>

A partir de ahí, se pueden realizar acciones sobre una tarea determinada. Por ejemplo, si quiere obtener detalles adicionales sobre la misma, puede hacer clic en este botón **`Instance Details`**:

<center><img src="https://i.postimg.cc/FRvX571N/a72.png"></center>

Y como puedes ver, obtienes información adicional como el DAG ID del DAG al que pertenece la tarea, el DAG Run, la duración de la tarea, etc.

<center><img src="https://i.postimg.cc/vTPJVhhZ/a73.png"></center>

Luego tengo el botón **`Rendered`**. Y este botón me permite saber cuál será la salida de los datos que inyecto en mi tarea en tiempo de ejecución (at run time). Esto está relacionado con el templado.

<center><img src="https://i.postimg.cc/rsF6s3SH/a74.png"></center>

En el botón de **`Logs`**, si hago clic en él se obtienen los logs de la tarea. Puede ver si la tarea está marcada como exitosa (success) o no.

<center><img src="https://i.postimg.cc/Y9xKpMmg/a75.png"></center>

<center><img src="https://i.postimg.cc/4NykgC47/a76.png"></center>

<center><img src="https://i.postimg.cc/QdK2FpTj/a77.png"></center>

El botón **`All instances`** (Todas las instancias) permite comprobar todas las instancias de una tarea determinada en todas las ejecuciones del DAG (DAG Runs). Por ejemplo, si abrimos esa vista, tenemos todas las instancias de la tarea batch_print_date1 que se han creado en todas las ejecuciones del DAG (DAG Runs).

<center><img src="https://i.postimg.cc/28tpjH3H/a78.png"></center>

<center><img src="https://i.postimg.cc/1zxLwfGw/a79.png"></center>

A continuación, tenemos las **`Task Actions`**. Por ejemplo, puedes ejecutar tu tarea si utilizas el executor adecuado, ya sea el CeleryExecutor o el KubernetesExecutor. Por ahora, no puedes hacer eso ya que usamos el LocalExecutor.

<center><img src="https://i.postimg.cc/s2pbcsm8/a80.png"></center>

A continuación, puedes Limpiar (**`Clear`**) tu tarea. Siempre que tenga una tarea en fallo (failure), solucione el problema y luego haga clic en "Clear" para reiniciar la tarea, con el fin de volver a intentar la tarea. Así que si haces clic en "Clear" ahora mismo y haces clic en "Ok" esta instancia de tarea (task instance) va a ser reintentada:

<center><img src="https://i.postimg.cc/FRpqMp2V/a81.png"></center>

<center><img src="https://i.postimg.cc/d1wpNpS0/a82.png"></center>

Como puede ver, el color del borde de la tarea es ahora blanco, lo que significa que todavía no hay ningún estado. Pero la cosa es que el estado de esa tarea no cambiará hasta que se active el toggle del DAG.

<center><img src="https://i.postimg.cc/HnLKd5mD/a83.png"></center>

Por lo tanto, tan pronto como actives el toggle para iniciar la programación (scheduling) de tus tareas. Como puedes ver bash_print_date1 se está ejecutando. Así que haciendo clic en "Clear", puede volver a intentar sus tareas.

<center><img src="https://i.postimg.cc/XN2mx2KV/a84.png"></center>

Por último, tiene "Mark Failed" y "Mark Success". Estos botones son útiles si quieres entender cómo se comportan las tareas que dependen de tu tarea según su estado.

<center><img src="https://i.postimg.cc/ncC5WvTq/a85.png"></center>
