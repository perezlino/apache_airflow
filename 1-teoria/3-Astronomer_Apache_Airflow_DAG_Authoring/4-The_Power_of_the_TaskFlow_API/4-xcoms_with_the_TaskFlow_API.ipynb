### **4.4 - XComs with the TaskFlow API**

Volvamos a la pregunta que teníamos en el vídeo sobre XComs. Si recuerdas tienes esta tarea "extract", donde quieres compartir múltiples valores con la tarea "process". Y así, con el fin de empujar (push) múltiples valores al mismo tiempo, has creado este diccionario con los valores "partner_name" y "partner_path". 

<center><img src="https://i.postimg.cc/ZRSgkZvd/a1200.png"></center>

Pero ahora, al devolver este diccionario, va a crear un XCom para todos sus valores. Pero qué pasa si quieres tener un XCom para un valor, un XCom para "partner_name" y un XCom para "partner_path", sin usar "xcom_push" dos veces. Pues bien, aquí es donde tienes que utilizar la Taskflow API. Una cosa que debes saber es que puedes pasar algunos argumentos a tus decoradores, por ejemplo, imaginemos que en lugar de tener el task ID por defecto correspondiente al nombre de tu función, quieres tener otro task ID. En ese caso, puedes pasar el siguiente argumento a tu decorador "task_ID=extract_partners". 

<center><img src="https://i.postimg.cc/267trJnX/a1201.png"></center>

Si guardas el archivo, y echas un vistazo a la interfaz de usuario de Airflow, haz clic en "my_dag" y luego ve a la Graph view, y como puedes ver, esta vez, el nombre de tu tarea es "extract_partners" como se especifica en el task ID.

<center><img src="https://i.postimg.cc/tJRMT55R/a1202.png"></center>

Ahora, ¿qué más puedes hacer? Bueno, si vuelves a tu editor de código, una cosa que puedes hacer es usar otro argumento, que es "multiple_outputs" igual a "True" y al especificar este argumento a true, estás diciendo que este XCom 

<center><img src="https://i.postimg.cc/8kwYY2p7/a1203.png"></center>

no debería ser un XCom con su diccionario como valor, sino que debería tener dos XComs, un XCom con la key "partner_name" y el valor "partner_name" y otro XCom con la key "partner_path" con el valor "partner_path". Así que, esta es la belleza de "multiple_outputs", con ella puedes decir, ¡hey! esto tiene múltiples outputs, múltiples XComs, que son empujados a la meta base de datos, así que no pongas todos esos valores en un solo XCom, sino pon esos valores en diferentes XComs y eso es exactamente lo que estás haciendo aquí. Así que, si guardas el archivo y ejecutas el DAG, veamos lo que obtienes. Vamos a activar el toggle del DAG, luego refrescamos la página, como puedes ver tenemos un DAG Run corriendo. Esperemos un poco y ya está hecho. Si vas a ADMIN > XComs

<center><img src="https://i.postimg.cc/6qH1Gvjr/a1204.png"></center>

cómo puedes ver obtienes dos XComs, uno correspondiente a "partner_name" con el valor "netflix" y otro "partner_path" con el valor "/partners/netflix". 

<center><img src="https://i.postimg.cc/rFmH2fsQ/a1205.png"></center>

Ahora, hay una cosa menos que puedes hacer, que es incluso mejor que usar ese argumento. Si vuelves a tu editor de código, otra forma de expresar que quieres poner varios valores en dos XComs diferentes, es eliminando "multiple_outputs" y especificando el tipo del valor devuelto por tu función Python. Como devuelves un diccionario, aquí 

<center><img src="https://i.postimg.cc/J422rjBm/a1206.png"></center>

defines "Dict[str, str]" y no te olvidas de importar el tipo dict, así que escribes "from typing import Dict" y esto es absolutamente equivalente que usar "multiple_outputs" igual a "True". Así que, esa es la belleza de la Taskflow API, realmente puedes manipular tus XComs mucho mejor que sólo usando x_push y xcom_pull y aquí eres capaz de empujar múltiples XComs a la vez, lo que está mejor optimizado. 

<center><img src="https://i.postimg.cc/dVZSLD8C/a1207.png"></center>

Usemos la otra notación con "multiple_outputs" y entonces la pregunta es, ¿cómo puedes obtener esos XComs en la tarea "process"? 

<center><img src="https://i.postimg.cc/s2FHSycN/a1208.png"></center>

Bueno, esto es super fácil, sólo tienes que añadir otro argumento, así que vamos a llamarlo "partner_path" entonces quieres imprimir "partner_path" también, y aquí, en lugar de pasar directamente "extract" como argumento de "process", creas una nueva variable, vamos a llamarla "partner_settings" igual a "extract( )", y entonces dices, quiero sacar el XCom con la key "partner_name" y luego quiero sacar el XCom con la key "partner_path". Y al hacer esto, usted es capaz de extraer (pull) sus XComs especificando la key como se muestra allí:

<center><img src="https://i.postimg.cc/Twfk79Bj/a1209.png"></center>

Es tan simple como eso. Vamos a comprobar si funciona. Así que vuelve a tu interfaz de usuario de Airflow. Activemos el toggle, actualicemos la página, como puede ver la ejecución del DAG se está ejecutando. Ok, ahora se ha completado, si vas a ADMIN > XComs, tienes tus XComs: "partner_name" y "partner_path". Si vuelves a tu DAG, en Graph view y haces click en la tarea "process", el click en "Log", como puedes ver obtienes "netflix" así como "/partners/netflix" como se define en los XComs.
