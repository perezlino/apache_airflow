### **5.2 - TaskGroups - The Best Way of Grouping your Tasks**

¿Recuerdas lo doloroso que era para agrupar tareas con "Subdags"? Bueno, esto se acabó con los "Task groups", es mucho más fácil porque no tendrás que lidiar con todos esos detalles técnicos, como asegurarte de que los "default arguments" están correctamente configurados con la "start date", o asegurarte de que el Subdag ID está correctamente definido con esta notación especial o incluso crear un nuevo Dag sólo para agrupar tus tareas. Es realmente mucho trabajo para hacer esto. Por eso se ha introducido su nuevo concepto en Airflow 2.0, el concepto de " task groups ". Vamos a sumergirnos en este hermoso mundo. La gran diferencia entre los "task groups" y los "subdags" es que con el " task group" usted agrupa sus tareas visualmente. No tienes que poner un Dag dentro de otro Dag y tus tareas en él. No tienes que hacer eso. No hay grandes cosas técnicas que hacer sólo para agrupar sus tareas. Así que vamos a ver cómo crear su primer "task group". Y para ello necesitas importar ese objeto, así que escribe "s".
